```python
# Importação do Pandas 

import pandas as pd


# Separando a url para o pandas buscar a tabela no site escolhido

url = r'https://pt.wikipedia.org/wiki/Lista_de_presidentes_do_Brasil'
pres = pd.read_html(url, match = 'Júlio')
Tabela_Pres = pres[0]


#Limpeza de dados irrelevântes para a minha análise -  Fotografias, notas e referências 
Tabela_Pres.drop('Referências e notas',axis = 1, inplace = True)
Tabela_Pres.drop('Fotografia',axis = 1,inplace = True)

# A Tabela é dividida em Repúblicas- da primeira até a sexta - Então eu dividi a tabela em várias, tornando as repúblicas como headers das colunas.

#Tabela 1
Tabela_Pres1 = Tabela_Pres[:17]
# display(Tabela_Pres1)

#Tabela 2

headers2 = Tabela_Pres.iloc[17]
Tabela_Pres2 = pd.DataFrame(Tabela_Pres.values[18:21], columns = headers2)

# display(Tabela_Pres2)

#Tabela3

headers3 = Tabela_Pres.iloc[21]
Tabela_Pres3 = pd.DataFrame(Tabela_Pres.values[22:24],columns = headers3)
# display(Tabela_Pres3)

#Tabela4

headers4 = Tabela_Pres.iloc[24]
Tabela_Pres4 = pd.DataFrame(Tabela_Pres.values[25:33],columns = headers4)
# display(Tabela_Pres4)

#Tabela5 

headers5 = Tabela_Pres.iloc[34]
Tabela_Pres5 = pd.DataFrame(Tabela_Pres.values[35:43],columns = headers5)
# display(Tabela_Pres5)

#Tabela6
headers6 = Tabela_Pres.iloc[44]
Tabela_Pres6 = pd.DataFrame(Tabela_Pres.values[45:],columns = headers6)
# display(Tabela_Pres6)


# Agora eu preciso transformar essas tabelas em um arquivo no excel com várias abas


with pd.ExcelWriter('Todos_os_Presidentes.xlsx') as writer:  
    Tabela_Pres1.to_excel(writer, sheet_name='1º República')
    Tabela_Pres2.to_excel(writer, sheet_name='2º República')
    Tabela_Pres3.to_excel(writer, sheet_name='3º República')
    Tabela_Pres4.to_excel(writer, sheet_name='4º República')
    Tabela_Pres5.to_excel(writer, sheet_name='5º República')
    Tabela_Pres6.to_excel(writer, sheet_name='6º República')
```
